```python
import pandas as pd

print("Loading IMDb datasets...")

# Load datasets
basics = pd.read_csv("data/title.basics.tsv", sep="\t", low_memory=False)
akas = pd.read_csv("data/title.akas.tsv", sep="\t", low_memory=False)

print("Files loaded")

# Keep only movies
basics = basics[basics["titleType"] == "movie"]

# Bollywood condition:
# Region = IN (India) AND language = hi (Hindi)
bollywood_ids = akas[
    (akas["region"] == "IN") & (akas["language"] == "hi")
]["titleId"].unique()

# Filter Bollywood movies from basics
bollywood_movies = basics[basics["tconst"].isin(bollywood_ids)]

# Keep only useful columns
bollywood_movies = bollywood_movies[["primaryTitle", "genres", "startYear"]]

# Clean titles
bollywood_movies["primaryTitle"] = bollywood_movies["primaryTitle"].str.strip()

# Save final Bollywood dataset
bollywood_movies.to_csv("data/bollywood_movies_imdb.csv", index=False)

print("âœ… Bollywood dataset created successfully!")
print("Total Bollywood movies:", len(bollywood_movies))
```
